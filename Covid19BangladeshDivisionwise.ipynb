## Proper Date Format

```python
# convert "Date" column to datetime
# TO DO : Parse month-year with a placeholder day (01)
# TO DO : Create sequential day number within each month-year group
# TO DO : Construct actual date by adding the offset
# TO DO : Remove the temporary columns


# parse month-year and store it in BaseDate column
covid['BaseDate'] = pd.to_datetime(covid['Date'].str.strip(), format='%b-%y', errors='coerce')

# create sequential day number within the month-year group and store it in DayOffset column
covid['DayOffset'] = covid.groupby('BaseDate').cumcount()

# Actual date by adading the date offset
covid['Date'] = covid['BaseDate'] + pd.to_timedelta(covid['DayOffset'], unit='D')

# drop temporary columns
covid.drop(columns=['BaseDate', 'DayOffset'], inplace=True)
```

## Handling Missing Value

### Observation (According to Histogram)
1. Most columns are right-skewed, meaning the majority of the data is concentrated toward the lower end (left side), and there's a long tail to the right.
2. In many histograms, there's a large spike at 0, which likely reflects your NaN-filled-with-zero strategy.
3. Some histograms show discrete steps, indicating categorical-like or integer-discrete data.

### Decision
If the missing values truly represent the absence of cases or deaths (indicates case recorded), we can fill the missing values with 0. But filling missing values with 0 will distorts the analysis if the column indicates the ratio / rate.

### Action
- fill columns like Confirmed cases or Deaths, where 0 = no event.
- Use interpolation for smoother trends in rate columns (Case rate, Death rate). 

```python
# Interpolate missing values for numeric columns
# Fill NaN values in "Rate" columns with linear interpolation
for col in covid.columns:
    if col == "Date":
        continue                                                        # Skip the "Date" column

    elif "Rate" in col:
        covid[col] = covid[col].interpolate(method='linear')            # Check if the column is numeric and contains "Rate"
    
    else:
        covid[col] = covid[col].fillna(0)                               # For other numeric columns, fill NaN with 0
```

## Outliers Treatment

### Observation
1. Almost all "Confirmed Cases" and "Deaths" columns across divisions show:
    - High number of outliers
    - Right-skewed distribution (Median near bottom, long upper whisker)
    - Some features with extremely large spikes
2. Death Rate and Case Rate columns also contain:
    - Moderate to high outliers
    - Some variation among divisions

### Decision
When data contains valuable but extreme values, extreme outliers can be replaced with Capping (Winsorization). upper/lower bounds (Q1 - 1.5Ã—IQR or Q3 + 1.5Ã—IQR)

### Action
- IQR (Interquartile Range) method for columns with integer values (case and death counts)
- Z-Score method for float columns (rate columns).

```python
# implement function for IQR method
def wisker(col):
    q1, q3 = np.percentile(col, [25, 75])
    iqr = q3 - q1
    lower_wisker = q1 - 1.5 * iqr
    upper_wisker = q3 + 1.5 * iqr
    return lower_wisker, upper_wisker
```

```python
# implement function for Z-score method
def z_score(col, threshold=3):
    mean = col.mean()
    std_dev = col.std()
    lower_wisker = mean - threshold * std_dev
    upper_wisker = mean + threshold * std_dev
    return lower_wisker, upper_wisker
```

```python
for col in covid.columns[1:]:
    if "Rate" not in col:
        lower_wisker, upper_wisker = wisker(covid[col])
        covid[col] = np.where(covid[col] < lower_wisker, lower_wisker, covid[col])
        covid[col] = np.where(covid[col] > upper_wisker, upper_wisker, covid[col])
    
    else:
        lower_wisker, upper_wisker = z_score(covid[col])
        covid[col] = np.where(covid[col] < lower_wisker, lower_wisker, covid[col])
        covid[col] = np.where(covid[col] > upper_wisker, upper_wisker, covid[col])
```

## Unipivot the clean dataset

```python
# reading the new dataset
df = pd.read_csv("files/covid_dataset_cleaned.csv")
```

```python
df.head()
```

```python
# exclude 'Date' column for melting
value_vars = [col for col in df.columns if col != 'Date']
```

```python
# Melt (unpivot)
df_long = df.melt(id_vars=['Date'], value_vars=value_vars, var_name='Metric', value_name='Value')

df_long.head()
```

```python
# split 'Metric' into 'Region' and 'Type'
df_long[['Type', 'Region']] = df_long['Metric'].str.extract(r'(.+?)\s+(Barisal|Chittagong|Dhaka|Khulna\
                                                            |Mymensingh|Rajshahi|Rangpur|Sylhet|Allover Bangladesh)', expand=True)
```

```python
# final clean up
df_long.drop(columns=['Metric'], inplace=True)

# reorder columns
df_long = df_long[['Date', 'Region', 'Type', 'Value']]
```
