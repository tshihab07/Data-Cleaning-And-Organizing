# Cleaning Customer Call List

### Tasks
   - drop duplicates
   - Clean unwanted non-alphanumeric(symbols)
   - Standardizing phone numbers
   - Multiple columns creation based on one single column
   - Remove null value
   - drop unwanted rows

````python
import pandas as pd
````

````python
# reading excel file
df = pd.read_excel(r"files\customer-call-list.xlsx")
df
````

````python
# dropping duplicate rows
df = df.drop_duplicates()
df
````

**NOTE:** dropping duplicates is the first stage of cleaning.

````python
# dropping unuseful column
df = df.drop('Not_Useful_Column', axis=1)
df
````

**NOTE:** Dropping unwanted columns make the dataset smaller.

````python
# cleaning names that have garbage with them
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
````

**NOTE:** strip() function is used to remove any leading, and trailing unwanted symbol. In our case, removes any leading or trailing characters **`'1', '2', '3', '.', '_', or '/'`** from the **`Last_Name`** column.

**`It is not necessary to match the whole pattern, but also it removes leading and trailing characaters if they are present in the argument`**

````python
# Task : Standardizing Phone numbers
# TO DO : remove non-alphanumeric characters
# TO DO : convert column values into string
# TO DO : Format Numbers

# removes non-alphanumeric and convert each value into string
phone_col = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", "", regex=True).astype(str)

# formating phone numbers
df["Phone_Number"] = phone_col.apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}")

df
````

**NOTE:**
   - ***str.replace():*** Replace non-alphanumeric characters with an empty space using a regular expression.
   - ***astype(str):*** Converts the column values to strings in one step.
   - ***apply():*** Formats the phone numbers with hyphens.

**Similar Code:**
   - Step1: Replace non-alphanumeric characters (Symbols)
   - Step2: Convert column values into string
   - Step3: Format the phone numbers

```python
# removing symbols(non-alphanumeric) inside the numbers
df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", "")

# converting column values into string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# formating numbers
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[:3] + "-" + x[3:6] + "-" + x[6:])
```

**One Line Approach:**
```python
df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", "", regex=True).astype(str).apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}")
```

````python
# cleaning null values of the phone number columns
df["Phone_Number"] = df["Phone_Number"].str.replace(r"(nan--|Na--)", "", regex=True)
df
````

**NOTE:** **`r"(nan--|Na--)"`** is a regular expression that matches either "nan--" or "Na--". If it matches then replace them with an empty string.

**Similar Code:**
```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "")
```

**One Line Approach:**
```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"(nan--|Na--)", "", regex=True)
```

````python
""" split column (create multiple columns based on one column) """

# creating new columns and assigining values
df[["Street", "State", "Zip_Code"]] = df["Address"].str.split(", ", 2, expand=True)

df
````

**NOTE:** split **`Address`** column into 3 division where the second argument indicate the max split. Then create three new columns and assign the splitted values in those columns

````python
# Task: standardizing multiple columns valu
# TO DO : Replace The values
df[["Paying Customer", "Do_Not_Contact"]] = df[["Paying Customer", "Do_Not_Contact"]].replace({"Yes":"Y", "No": "N"})
df
````

**NOTE:** We can replace multiple items by passing a dictionary in the replace() function. By accessing multiple columns, we can replce the same items from each column

**Similar Code:**
```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

**One Line Approach:**
```python
df[["Paying Customer", "Do_Not_Contact"]] = df[["Paying Customer", "Do_Not_Contact"]].replace({"Yes":"Y", "No": "N"})
```

````python
# clean up NaN data from the dataset
df = df.fillna("")
df
````

**NOTE:** we can not use **`replace()`** function to replace NaN values in the column. Because **`replace()`** is used to replace a string. Since NaN is not a string, we have to use the **`fillna()`**

````python
# filtering rows
for item in df.index:
    if df.at[item, "Do_Not_Contact"] == "Y":
        df.drop(item, inplace=True)

df
````

**NOTE:** We can use conditional selection technique to filter rows.

**Example:**
```python
df = df[df["Do_Not_Contact"] != "Y"]
```

````python
# filtering rows
df = df[df["Phone_Number"] != ""]
df
````

**NOTE:** Conditional Selection

**Similar Code:**
```python
for item in df.index:
    if df.at[item, "Phone_Number"] == "":
        df.drop(item, inplace=True)
```

````python
# final touch
df = df.reset_index(drop=True)
df
````

````python
df.to_excel(r"files\customer-call-list-cleaned.xlsx")
````
