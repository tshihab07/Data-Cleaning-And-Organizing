# Title: Cricket Wicket Fall Visualization
**Description:** This Python script uses Matplotlib to visualize the fall of wickets in a cricket match between Bangladesh and Zimbabwe. The plot displays the number of wickets fallen over each over bowled, with the x-axis representing overs and the y-axis representing runs.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
""" sample data """
# fall of wicket
bd_wicket_fall = np.array([0, 101, 108, 121, 122, 123, 130, 130, 132, 138, 143])
bd_overs = np.array([0, 11.2, 11.6, 13.4, 14.1, 14.6, 16.2, 16.5, 17.3, 18.3, 19.5])


# overs
zim_wicket_fall = np.array([0, 0, 28, 32, 57, 92, 94, 103, 128, 138, 138])
zim_overs = np.array([0, 0.4, 3.5, 4.4, 9.3, 14.3, 14.6, 16.1, 18.3, 19.3, 19.4])
```

```python
# Plotting
plt.figure(figsize=(10, 6))

plt.plot(bd_overs, bd_wicket_fall, marker='o', label='Bangladesh', color="green", mfc="r")
plt.plot(zim_overs, zim_wicket_fall, marker='o', label='Zimbabwe', color="red", mfc="y")

plt.xticks(np.linspace(0, 20, 5))
plt.yticks(np.linspace(0, 150, 7))

plt.title('Score Comparison')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.legend()
plt.grid(axis="y")
plt.show()
```
